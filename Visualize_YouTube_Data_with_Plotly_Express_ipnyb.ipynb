### Import libraries

```python
import pandas as pd
import io
```

### Import dataset

```python
df = pd.read_csv('most_subscribed_youtube_channels.csv')
display(df)
```

```python
for col in ['subscribers', 'video views', 'video count']:
  df[col] = df[col].astype(str).str.replace(',', '')
display(df)
```

### Making Histogram

```python
import plotly.express as px
fig = px.histogram(df, x='subscribers', title='Subscriber Count')
fig.show()
```

###Make a Pie Chart

```python
fig = px.pie(df, values='subscribers', names='category', title='YouTube Categories')
fig.show()
```

###Make a box plot

```python
fig = px.box(df, y='started', title='Years Started', points='all')
fig.show()
```
